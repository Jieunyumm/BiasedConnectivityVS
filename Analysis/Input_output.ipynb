# Input-Output Relationship 

```python
import pandas as pd
import numpy as np
import pdb
import seaborn as sns
from matplotlib import pyplot as plt
import os
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "browser"
```

```python
# Load dataset 

#Change to directory containing .csv file
file_dir = "/Users/jieunyum/Desktop/Y2/NEUR0024/extension_workshops/BiasedConnectivityVS/Wee_2020_TRIO"
os.chdir(file_dir)
 
df_LH = pd.read_csv("LH_TRIO_cellcounts.csv")
df_NAc = pd.read_csv("NAc_TRIO_cellcounts.csv")
df_PFC = pd.read_csv("PFC_TRIO_cellcounts.csv")
```

### 1 Overview of Input 

```python
# Concatenate into 1 data frame according to projection type 
df_LH2 = df_LH.copy()
df_NAc2 = df_NAc.copy()
df_PFC2 = df_PFC.copy()

df_LH2["Projection"] = "LH"
df_NAc2["Projection"] = "NAc"
df_PFC2["Projection"] = "PFC"

df_all = pd.concat([df_LH2, df_NAc2, df_PFC2], ignore_index=True)
```

```python
# Plot histogram 
fig = plt.plot() 

counts = df_all['Coarse'].value_counts()

plt.bar(counts.index, counts.values)
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")

plt.show()
```
